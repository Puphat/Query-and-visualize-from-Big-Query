```python
from google.colab import auth
from google.cloud import bigquery
auth.authenticate_user()
print('Authenticated')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def bq_run(query_text):
  project_id = 'madt-project'

  client = bigquery.Client(project=project_id)

  df = client.query(query_text).to_dataframe()

  return df
```

```python
#อัพเดตยอดขายวันนี้

query_text = '''
                SELECT 
                sum(sales) as Sales,
                FROM `madt-project.madt7101.superstore` 
                WHERE order_date = current_date()
                GROUP BY order_date
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#อัพเดตยอดขายเดือนนี้

query_text = '''
                SELECT  
                sum(Sales) as Sales,
                FROM (SELECT 
                sum(sales) as Sales,
                order_date,
                1 as Index
                FROM `madt-project.madt7101.superstore` 
                WHERE order_date <= current_date AND order_date >= datetime_trunc(current_date, month)
                GROUP BY order_date)
                GROUP BY Index
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#พื้นที่ขายดีที่สุด

query_text = '''
                SELECT region
                FROM (SELECT 
                sum(sales) as Sales,
                region
                FROM `madt-project.madt7101.superstore` 
                GROUP BY region
                ORDER BY Sales DESC
                LIMIT 1)
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#ประเภทสินค้าที่ขายดี

query_text = '''
                SELECT sub_category
                FROM (SELECT 
                sum(sales) as Sales,
                sub_category
                FROM `madt-project.madt7101.superstore` 
                GROUP BY sub_category
                ORDER BY Sales DESC
                LIMIT 1)
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#อัพเดทต้นทุนเดือนนี้

query_text = '''
                SELECT  
                sum(cost) as Cost,
                FROM (SELECT 
                sum(cost) as Cost,
                order_date,
                1 as Index
                FROM `madt-project.madt7101.superstore` 
                WHERE order_date <= current_date AND order_date >= datetime_trunc(current_date, month)
                GROUP BY order_date)
                GROUP BY Index
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#พื้นที่ที่มีต้นทุนของสินค้าสูงที่สุด

query_text = '''
                SELECT region
                FROM (SELECT 
                sum(cost) as Cost,
                region
                FROM `madt-project.madt7101.superstore` 
                GROUP BY region
                ORDER BY Cost DESC
                LIMIT 1)
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#ประเทศที่มีต้นทุนของส่งสินค้าสูงที่สุด

query_text = '''
                SELECT country
                FROM (SELECT 
                sum(cost) as Cost,
                country
                FROM `madt-project.madt7101.superstore` 
                GROUP BY country
                ORDER BY Cost DESC
                LIMIT 1)
             '''
#Create Data Frame
df = bq_run(query_text)
df
```

```python
#สินค้าที่มีต้นทุนสูงที่สุด

query_text = '''
                SELECT sub_category
                FROM (SELECT 
                sum(cost) as Cost,
                sub_category
                FROM `madt-project.madt7101.superstore` 
                GROUP BY sub_category
                ORDER BY Cost DESC
                LIMIT 1)
             '''
#Create Data Frame
df = bq_run(query_text)
df
```
